# Optimización escalar con SymPy: criterio conocido

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.

1. $f(x)=x+\frac{1}{x^2}$

```python
#Definir variables
x=sp.var("x",real=True)
#Definimos función
def f(x):
  return x+1/x**2
#Derivamos
df=f(x).diff()
#Resolvemos
c=sp.solve(df,x)
c
#Derivamos nuevamente para máximo o mínimo
d2f=df.diff()
d2f
#Graficamos
f_num=sp.lambdify(x,f(x),"numpy")
x_num=np.linspace(0,10,100)
y_num=f(x_num)
plt.plot(x_num,y_num)
plt.plot(float(c[0]),f_num(float(c[0])),'ro',ms=6,label=f"{c[0]:.2f}")
plt.legend()
plt.grid()
plt.show()
```

2. $f(x)=1260+117x-9x^2$

El criterio de la segunda derivada nos arroja un mínimo global donde $f'(1.26)=0$

```python
#Definimos variables
x=sp.var("x",real=True)
#Definimos la función
def f(x):
  return 1260+117*x-9*x**2
#Derivamos
df=f(x).diff()
#Resolvemos
c=sp.solve(df,x)
c
#Derivamos nuevamente para máximo o mínimo
d2f=df.diff()
d2f
#Graficamos
f_num=sp.lambdify(x,f(x),"numpy")
x_num=np.linspace(0,10,100)
y_num=f(x_num)
plt.plot(x_num,y_num)
plt.plot(float(c[0]),f_num(float(c[0])),'ro',ms=6,label=f"{c[0]:.2f}")
plt.legend()
plt.grid()
plt.show()
```

El criterio de la segunda derivada nos arroja un máximo global de $f'(6.50)=0$

3. $f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
#Definimos variables
x=sp.var("x",real=True)
#Definimos la función
def f(x):
  return 5+x+4/x
#Derivamos
df=f(x).diff()
#Resolvemos
c=sp.solve(df,x)
c
#Derivamos nuevamente para máximos o mínimos
d2f=df.diff()
d2f
#Graficamos y dado que el dominio [x > 0], entonces tomamos los valores de C que sean mayores a 0
f_num=sp.lambdify(x,f(x),"numpy")
x_num=np.linspace(0,10,100)
y_num=f(x_num)
plt.plot(x_num,y_num)
for i in c:
  if i>0:
    plt.plot(i,f_num(i),'ro',ms=6,label=f"{i:.2f}")
plt.legend()
plt.grid()
plt.show()
```

El criterio de la segunda derivada nos arroja un valor mínimo global, debido a que analizamos el dominio donde $x>0$ sólo tomamos el valor de la primera derivada que cumpla con este requisito. Donde $f(2)=0$.

4. $f(x)=\frac{15x}{2}\sqrt{\frac{x}{x-30}}$

```python
#Definimos variables
x=sp.var("x",real=True)
#Definimos la función
def f(x):
  return 15*x/2*((x/(x-30))**0.5)
#Derivamos
df=f(x).diff()
#Resolvemos
c=sp.solve(df,x)
c
#Derivamos nuevamente para máximos o mínimos
d2f=df.diff()
d2f
#Graficamos con ambos puntos
f_num=sp.lambdify(x,f(x),"numpy")
x_num=np.linspace(-50,50,10000)
y_num=f(x_num)
plt.plot(x_num,y_num)
for i in c:
  plt.plot(i,f_num(i),'ro',ms=6,label=f"{i:.2f}")
plt.legend()
plt.grid()
plt.show()
```

El criterio de la segunda derivada nos arroja dos valores globales: un límite de dominimo y un mínimo, donde $f'(45)=0$ y $f'(0)=0$. Aquí el domino aparece como $x∈(-∞,0] 𝑈 (30+∞)$

5. $f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
#Definimos variables
x=sp.var("x",real=True)
#Definimos la función
def f(x):
  return 6-2*x+((18-2*x**2)**0.5)
#Derivamos
df=f(x).diff()
#Resolvemos
c=sp.solve(df,x)
c
#Derivamos nuevamente para máximos y mínimos
d2f=df.diff()
d2f
#Graficamos
f_num=sp.lambdify(x,f(x),"numpy")
x_num=np.linspace(-3,3,10000)
y_num=f(x_num)
plt.plot(x_num,y_num)
plt.plot(float(c[0]),f_num(float(c[0])),'ro',ms=6,label=f"{c[0]:.2f}")
plt.legend()
plt.grid()
plt.show()
```

El criterio de la segunda derivada nos arroja un máximo global donde $f'(6.50)=0$

6. $f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
#Definimos variables
x=sp.var("x",real=True)
#Definimos la función
def f(x):
  return 1/4*x**4+1/3*x**3-x**2
#Derivamos
df=f(x).diff()
#Resolvemos
c=sp.solve(df,x)
c
#Derivamos nuevamente para máximos o mínimos
d2f=df.diff()
d2f
#Graficamos con todos los puntos
f_num=sp.lambdify(x,f(x),"numpy")
x_num=np.linspace(-3,1.7,10000)
y_num=f(x_num)
plt.plot(x_num,y_num)
for i in c:
  plt.plot(i,f(i),'ro',ms=6,label=f"{i:.2f}")
plt.legend()
plt.grid()
plt.show()
```

La derivada nos arroja tres valores: dos mínimos (local y global) y un máximo global. Dentro del rango [-3,1.7], -2 y 0 son los valores globales y 0 es el local.

7. $f(x)=100(1+\cos(x)\sin(x))$, para $x\in[0.\frac{\pi}{2}]$

```python
#Definimos variables
x=sp.var("x",real=True)
#Definimos función
def f(x):
  return 100*(1+sp.cos(x)*sp.sin(x))
#Derivamos
df=f(x).diff()
#Resolvemos
c=sp.solve(df,x)
c
#Graficamos con todos los puntos críticos
f_num=sp.lambdify(x,f(x),"numpy")
x_num=np.linspace(0,(np.pi/2),10000)
y_num=f_num(x_num)
plt.plot(x_num,y_num)
for i in c:
  if i>0 and i<(np.pi/2):
    plt.plot(i,f(i),'ro',ms=6,label=f"{i:.2f}")
plt.legend()
plt.grid()
plt.show()
```

La derivada arroja dos valores globales, pero como el dominio es de $[0,π/2]$ entonces solo agarramos $f'(0.79)=0$, el cual es un máximo global
